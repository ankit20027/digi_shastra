```python
import pandas as pd
import numpy as np
import json
```

```python
data = pd.read_excel('../bhagavad-gita.xlsx', sheet_name=None)
```

## preface

```python
data['Adhyay Details']
```

```python
preface = data['Adhyay Details'].loc[:, 'CHAPTER':'NO. OF SHLOKAS']
preface = preface.set_index('CHAPTER')
preface = preface[:-1]
json.dump(obj=preface.T.to_dict('dict'), fp=open('preface.json', 'w+'), indent=4)
```

## shlokas

```python
data['Bhagavad-Gita']['Chapter'] = data['Bhagavad-Gita']['Chapter'].apply(lambda x: x.split(' ')[1])
data['Bhagavad-Gita']['Verse'] = data['Bhagavad-Gita']['Verse'].apply(lambda x: x.split('.')[1])
data['Bhagavad-Gita'].drop(data['Bhagavad-Gita'].columns[:2], axis=1, inplace=True)
data['Bhagavad-Gita']
```

```python
data['Bhagavad-Gita'].to_csv('bhagwat_geeta.csv', index=False)
```
